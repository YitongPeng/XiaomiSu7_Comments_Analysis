```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
file_path = r"D:\MINE\大数据分析综合实验\大数据2102班_彭弋桐_基于小米su7评论的情感分析和数据分析\数据\情感赋值.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# 删除 '内容' 列中包含空值的所有行
df = df.dropna()

# 删除重复
df = df.drop_duplicates()
df.head()
```

```python
# # 将日期列转换为日期时间类型
# df['日期'] = pd.to_datetime(df['日期'])

# # 计算与参考日期（2024/3/30 12:22）的分钟差
# reference_date = pd.Timestamp('2024-03-30 12:22')
# df['分钟差'] = (df['日期'] - reference_date) / pd.Timedelta(minutes=1)  # 将时间差转换为分钟

# df.head()
```

```python
# 设置字体属性
font_path = "C:\\Windows\\Fonts\\SimHei.ttf"  # Windows路径
font_prop = FontProperties(fname=font_path)


# 将日期列转换为日期时间类型
df['日期'] = pd.to_datetime(df['日期'])

# 按日期排序
df.sort_values(by='日期', inplace=True)

# 选择日期和评论数列
date_comments = df[['日期', 'length']]

# 按日期对评论数进行汇总
date_comments = date_comments.groupby('日期').sum()

# 绘制趋势图
plt.figure(figsize=(10, 6))
plt.plot(date_comments.index, date_comments['length'], linestyle='-')
plt.title('评论数量随时间的变化趋势', fontproperties=font_prop)
plt.xlabel('日期', fontproperties=font_prop)
plt.ylabel('评论数量', fontproperties=font_prop)
plt.grid(True)
plt.show()
```

```python
# 设置字体属性
font_path = "C:\\Windows\\Fonts\\SimHei.ttf"  # Windows路径
font_prop = FontProperties(fname=font_path)


# 筛选出4月1号到4月15号的数据
date_comments_april = date_comments[(date_comments.index >= '2024-04-01') & (date_comments.index <= '2024-04-15')]

# 绘制趋势图
plt.figure(figsize=(10, 6))
plt.plot(date_comments_april.index, date_comments_april['length'], linestyle='-')
plt.title('4月1号到4月15号评论数量随时间的变化趋势', fontproperties=font_prop)
plt.xlabel('日期', fontproperties=font_prop)
plt.ylabel('评论数量', fontproperties=font_prop)
plt.grid(True)
plt.show()
```
